```python
def seriation(children):
    N = children.shape[0] + 1  # Total number of elements or clusters
    stack = [2 * N - 2]  # Initialize stack with the root node index
    res_order = []

    while stack:
        cur_idx = stack.pop()
        if cur_idx < N:  # Original element
            res_order.append(cur_idx)
        else:  # Merged cluster
            child_1, child_2 = children[cur_idx - N].astype(int)
            stack.append(child_1)
            stack.append(child_2)

    return res_order[::-1]
```

```python
def recursiveBisection(V, l, r, W):
    #Performs recursive bisection weighting for a new portfolio

    #V is the sorted correlation matrix
    #l is the left index of the recursion
    #r is the right index of the recursion
    #W is the list of weights

    if r-l == 1: #One item
        return W
    else:
        #Split up V matrix
        mid = l+(r-l)//2
        V1 = V[l:mid, l:mid]
        V2 = V[mid:r, mid:r]

        #Find new adjusted V
        V1_diag_inv = 1/np.diag(V1)
        V2_diag_inv = 1/np.diag(V2)
        w1 = V1_diag_inv/V1_diag_inv.sum()
        w2 = V2_diag_inv/V2_diag_inv.sum()
        V1_adj = w1.T@V1@w1
        V2_adj = w2.T@V2@w2

        #Adjust weights
        a2 = V1_adj/(V1_adj+V2_adj)
        a1 = 1-a2
        W[l:mid] = W[l:mid]*a1
        W[mid:r] = W[mid:r]*a2
        W = recursiveBisection(V, l, mid, W)
        W = recursiveBisection(V, mid, r, W)
        return W
```
